## Vector addition

```python
v1 = np.array([4, 4])
v2 = np.array([1, 3])
product = v1 + v2

print(v1)
print("+")
print(v2)
print("=")
print(product)


plot_vectors(
    (v1, "v1"), 
    (v2, "v2"),
    (product, "product")
)
```

# Vector multiplication
## Scalar - vector multiplication

```python
stretched = v1 * 2
print(v1)
print("*")
print(2)
print("=")
print(stretched)

flipped = v1 * -0.3
print(v1)
print("*")
print(-0.3)
print("=")
print(flipped)

plot_vectors(
    (v1, "v1"), 
    (stretched, "v1 * 2"),
    (flipped, "v1 * -1.5")
)
```

# Transformation tests

```python

x_ax = np.array([10, 0])
y_ax = np.array([0, 10])

v1 = np.array([3, 4])
pretty_print_matrix(v1, "v1")

identity_matrix = np.array([[1, 0], [0, 1]])
identity_transformed = v1 @ identity_matrix
pretty_print_matrix(identity_transformed, "v1 after identity transformation")

transformation_matrix = np.array([[1, 0.5], [1.5, 1]])
transformation = v1 @ transformation_matrix
pretty_print_matrix(transformation, "v1 after a modifying transformation")

plot_vectors(
    (x_ax, "x axis"),
    (y_ax, "y axis"),
    (identity_transformed, "identity_transformed"),
    (transformation, "transformation"),
    (v1, "v1")
)
```

# Inverse matrices
> The inverse of a matrix is a matrix that when multiplied with the original matrix produces the identity matrix. It therefore compensates the transformation effect of the original matrix. The inverse of a matrix is denoted as $A^{-1}$.

```python
x_ax = np.array([10, 0])
y_ax = np.array([0, 10])

v1 = np.array([3, 4])
pretty_print_matrix(v1, "v1")

transformation_matrix = np.array([[1, 0.5], [1.5, 1]])
pretty_print_matrix(transformation_matrix, "transformation_matrix")
transformation = v1 @ transformation_matrix
pretty_print_matrix(transformation, "v1 after a modifying transformation")

inverse_transformation_matrix = np.linalg.inv(transformation_matrix)
pretty_print_matrix(inverse_transformation_matrix, "inverse_transformation_matrix")
inverse_transformation = transformation @ inverse_transformation_matrix
pretty_print_matrix(inverse_transformation, "inverse_transformation")

inverse_applied_to_v1 = v1 @ inverse_transformation_matrix
pretty_print_matrix(inverse_applied_to_v1, "inverse_applied_to_v1")

plot_vectors(
    (x_ax, "x axis"),
    (y_ax, "y axis"),
    (transformation, "transformation"),
    (inverse_transformation, "inverse_transformation"),
    (v1, "v1"),
    (inverse_applied_to_v1, "inverse_applied_to_v1")
)
```
